```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import (Audio, display, clear_output)
%matplotlib inline
```

```python
rate = 16000.
duration = .25
t = np.linspace(0., duration, int(rate * duration))
```

```python
def synth(f):
    x = np.sin(f * 2. * np.pi * t)
    display(Audio(x, rate=rate, autoplay=True))
    return x
```

```python
s=synth(440)
```

```python
plt.plot(t,s)
```
